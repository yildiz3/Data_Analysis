# Matplotlib: Legend Options

```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 1, 4, 6, 4])
```

## Add a Legend

### Basic Legend

```python
plt.plot(x, y)
plt.plot(x, y+2, color='orange')
plt.legend(['blue', 'orange']);
```

### `label` Argument

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend();
```

### Reorder Legend

```python
line_1, = plt.plot(x, y, label='blue')
line_2, = plt.plot(x, y+2, color='orange', label='orange')
plt.legend(handles=[line_2, line_1]);  
```

## Legend Location

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(loc='lower right');
```

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(loc='upper center');
```

The default `loc` location is "best", which selects the location (out of the nine basic locations) with the minimum overlay with the artists drawn on the figure.  This option can be slow if there are many elements drawn on the figure.

```python
plt.plot(x, 10-y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(loc='best'); 
```

### Outside Figure

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left');
```

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(bbox_to_anchor=(-0.05, 0), loc='lower right'); 
```

### Horizontal Legend (`ncol`)

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(ncol=2); 
```

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.plot(x, y+4, color='green', label='green')
plt.legend(ncol=3); 
```

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.plot(x, y+4, color='green', label='green')
plt.legend(ncol=3, bbox_to_anchor=(0.5, -0.1), loc='upper center'); 
```

## Styling

### Font Size

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(fontsize=14);
```

### Other Font Properties

#### Bold, Font Family (`prop`)

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(prop={'weight': 'bold', 'size': 14, 'family': 'serif'});
```

#### Font Color (`labelcolor`)

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(labelcolor=['blue', 'orange']);
```

### Legend Border (`frameon`)

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(frameon=False);
```

### Legend `title`

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(title='line colors');
```

### Legend Background Color (`facecolor`)

```python
plt.plot(x, y, label='blue')
plt.plot(x, y+2, color='orange', label='orange')
plt.legend(facecolor='lightgray');
```

## Thumbnail Image

```python
plt.figure(figsize=(6,4.5))
plt.plot(x, y, lw=3, label='blue')
plt.plot(x, y+2, lw=3,color='orange', label='orange')
plt.plot(x, y+4, lw=3,color='green', label='green')
plt.legend(bbox_to_anchor=(-0.1, 0.5), loc='center right',
           fontsize=14
          )
plt.xticks([1, 2, 3, 4, 5], fontsize=14)
plt.yticks(fontsize=14)
sns.despine()
#plt.savefig('08legend_thumb.png', dpi=200, bbox_inches='tight'); 
```
