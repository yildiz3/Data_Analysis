# Matplotlib: Adding Shading

```python
from matplotlib import pyplot as plt

import numpy as np
import seaborn as sns
```

```python
x = np.random.normal(10, 1, size=50)
y = .5*x + np.random.normal(5, 1, size=50)
```

## Add Shading

### Basic Vertical: `axvspan`

```python
plt.scatter(x, y)
plt.axvspan(10, 13);
```

#### `ymin`, `ymax`

```python
plt.scatter(x, y)
plt.axvspan(10, 13, ymin=0.5, ymax=0.75);
```

### Basic Horizontal: `axhspan`

```python
plt.scatter(x, y)
plt.axhspan(11, 14);
```

#### `xmin`, `xmax`

```python
plt.scatter(x, y)
plt.axhspan(11, 14, xmin=0.25, xmax=0.5); 
```

## Styling

```python
import pandas as pd
from datetime import datetime
```

```python
dates = [datetime(2019, 11, 15), datetime(2019, 12, 15), datetime(2020, 1, 15),
         datetime(2020, 2, 15), datetime(2020, 3, 15)]
prices = [110.40, 125.22, 78.30, 66.90, 78.05]
```

```python
dates
```

```python
plt.figure(figsize=(6, 4))

pd.Series(prices, index=dates).plot(marker='o', markersize=10)

plt.xticks([datetime(2019, 12, 1), datetime(2020, 1, 1), 
            datetime(2020, 2, 1),datetime(2020, 3, 1)], 
           ['Dec\n2019', 'Jan\n2020', 'Feb\n2020', 'March\n2020'], 
           rotation=0, 
           ha='center'
          ) 
plt.xlim(datetime(2019, 11, 10), datetime(2020, 3, 20))
plt.ylim(50, 150)

#add shading
plt.axvspan(datetime(2019, 11, 10), datetime(2020, 1, 1), color='lightgray', alpha=0.5)

#add text: 2019, 2020
plt.text(datetime(2019, 12, 15), 140, '2019', fontsize=16, color='gray')
plt.text(datetime(2020, 1, 5), 140, '2020', fontsize=16, color='black')

sns.despine();
```

### Variables and Polygon Properties

```python
plt.scatter(x, y)

x_mean = np.mean(x)

x_start, x_end = plt.xlim()

#add shading with pattern
p = plt.axvspan(x_mean, x_end, alpha=0.2)
p.set_hatch('/')

plt.xlim(None, x_end);
```

```python
type(p)
```

## Curves: `fill_between`

```python
x_val = np.linspace(-5, 5, 1000)
y_val = 20 - x_val**2
```

```python
plt.fill_between(x_val, -5, y_val);
```

```python
y1 = np.sin(0.5*x_val)
y2 = -3 + np.sin(-0.5*x_val)
```

```python
plt.fill_between(x_val, y1, y2);
```

```python
plt.fill_between(x_val, y1, y2, alpha=0.5, hatch='|');
```
