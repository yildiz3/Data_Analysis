# Matplotlib: Adding a Baseline

```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
X = np.linspace(0, 10, 1000)

Y = np.sin(np.pi*X/2) + 2
Y2 = 10 - (X-5)**2
```

## axhline, axvline

Draw a horizontal line with pyplot's `axhline` or a vertical line with `axvline`.

```python
plt.plot(X, Y)
plt.axhline(2);
```

```python
plt.plot(X, Y2)
plt.axvline(5, ymin=0.2, ymax=.8);
```

## Example Baselines

```python
import seaborn as sns
```

```python
plt.rcParams['font.size'] = 14
```

```python
students = ['A', 'B', 'C', "D", "E"]
grades = [80, 42, 68, 91, 77]

plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')
plt.axhline(100, color='xkcd:brick red', lw=3)
plt.text(2.5, 104, 'Maximum Grade', 
         fontstyle='italic',
         color='xkcd:brick red', fontweight='bold', fontsize=14)
for i,g in enumerate(grades):
    plt.text(i-0.18, g-8, f'{g}', color='black', fontsize=16)
    
plt.xlabel('Student')
sns.despine()
plt.tight_layout()
#plt.savefig('05baseline_fig_grades.png', dpi=200);
```

#### Used for Video Thumbnail

```python
students = ['A', 'B', 'C', "D", "E"]
grades = [80, 42, 68, 91, 77]

plt.figure(figsize=(5.5,7))
plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')
plt.axhline(100, color='xkcd:brick red', lw=5)
plt.text(2, 102, 'Maximum Grade', 
         fontstyle='italic',
         color='xkcd:brick red', fontweight='bold', fontsize=16)
for i,g in enumerate(grades):
    plt.text(i-0.18, g-5, f'{g}', color='black', fontsize=16)
    
#plt.xlabel('Student')
sns.despine()
plt.tight_layout()
#plt.savefig('05baseline_fig_thumb.png', dpi=200);
```

```python
np.random.seed(44)
x = np.random.normal(2, 1, size=30)
y = np.random.normal(4, 1, size=30)
```

```python
plt.scatter(x, y, c='xkcd:bruise', edgecolor='black', s=100, zorder=10)
plt.axvline(np.mean(x), color='gray', lw=6,  zorder=1)
plt.xlim(0, 4.4)
plt.ylim(0, 6.5)
plt.yticks([0, 2, 4, 6])
plt.xticks([0, 1, 2, 3, 4])
plt.text(0.45, 1, f'mean = {np.mean(x):.2f}', 
         fontstyle='italic',
         color='gray', fontweight='bold', fontsize=16)
sns.despine()
plt.tight_layout()
plt.savefig('05baseline_figure_vertMean.png', dpi=200);
```

```python
plt.scatter(x, y, c='xkcd:bruise', edgecolor='black', s=100, zorder=10)
plt.axhline(np.mean(y), color='gray', lw=6,  zorder=1)
plt.xlim(0, 4.4)
plt.ylim(0, 6.5)
plt.yticks([0, 2, 4, 6])
plt.xticks([0, 1, 2, 3, 4])
plt.annotate(f'mean = {np.mean(y):.2f}', xy=(2.7, 4), xytext=(3, 2),
             fontstyle='italic', color='gray', fontweight='bold', fontsize=16,
            arrowprops=dict(facecolor='gray', shrink=0.03, edgecolor='gray'))
sns.despine()
plt.tight_layout()
plt.savefig('05baseline_figure_horizMean.png', dpi=200);
```

## Line Styling

`axhline` and `axvline` inherit properties from matplotlib's [Line2D](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D).

### Color

```python
plt.plot(X, Y)
plt.axhline(2, color='orangered');
```

### Linewidth, `lw`

```python
plt.plot(X, Y)
plt.axhline(2, color='orangered', lw=4);
```

### Linestyle, `ls`

```python
plt.plot(X, Y)
plt.axhline(2, color='orangered', lw=4, ls='--');
```

### Layer order, `zorder`

```python
plt.plot(X, Y, lw=4, zorder=2)
plt.axhline(2, color='orangered', lw=4, zorder=1);
```

### label

```python
plt.plot(X, Y, lw=4, label='sine wave')
plt.axhline(2, color='orangered', lw=4, label='baseline')
plt.legend();
```
