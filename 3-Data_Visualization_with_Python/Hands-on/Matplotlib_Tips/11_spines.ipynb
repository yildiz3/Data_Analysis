# Matplotlib: Remove Spines

```python
from matplotlib import pyplot as plt
```

```python
x = range(10)
y = [0, 3, 5, 4, 5, 7, 8, 8, 7, 9]
```

## With matplotlib

```python
plt.plot(x, y)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
```

```python
type(ax)
```

```python
ax.spines
```

### Remove all spines with `set_frame_on`

```python
plt.plot(x, y)
ax = plt.gca()
ax.set_frame_on(False);
```

### Remove for all plots with `rcParams`

```python
plt.plot(x, y);
```

```python
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
```

```python
plt.plot(x, y);
```

```python
plt.scatter(x, y);
```

```python
plt.rcParams['axes.spines.top'] = True
plt.rcParams['axes.spines.right'] = True
```

## With Seaborn

```python
import seaborn as sns
```

```python
plt.plot(x, y);
```

### Removing other spines

By default, `sns.despine()` removes top and left spines.  Use arguments to update.

Default: `sns.despine(left=False, bottom=False, right=True, top=True)`

```python
plt.plot(x, y)
sns.despine();
```

```python
plt.plot(x, y)
sns.despine(left=True, right=False);
```

## Thumbnail Image and Overlays

### Thumbnail

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y, lw=5)
plt.xticks([])
plt.yticks([])
ax = plt.gca()
ax.spines['right'].set_color('gray')
ax.spines['right'].set_linestyle((0,(10,10)))
for spine in ax.spines:
    ax.spines[spine].set_linewidth(2)
ax.spines['right'].set_linewidth(1)
#plt.savefig('11spines_thumb.png', dpi=200, bbox_inches='tight'); 
```

### Overlays

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y, lw=5)
plt.xticks([])
plt.yticks([])
ax = plt.gca()
ax.set_frame_on(False)
#plt.savefig('11spines_noframe.png', dpi=200, bbox_inches='tight'); 
```

```python
spine_list = ['bottom', 'right', 'top', 'left']
for i in range(len(spine_list)):
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, lw=5)
    plt.xticks([])
    plt.yticks([])
    ax = plt.gca()
    for spine in spine_list[:i]:
        ax.spines[spine].set_visible(False)
    ax.spines[spine_list[i]].set_linewidth(5)
    ax.spines[spine_list[i]].set_color('darkorange')
    #plt.savefig(f'11spines_addspines{i}.png', dpi=200, bbox_inches='tight'); 
```
