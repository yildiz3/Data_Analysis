# Matplotlib: Adding Symbols with $\TeX$

```python
from matplotlib import pyplot as plt

import numpy as np
import matplotlib
```

```python
X = np.linspace(1, 4, 1000)
Y = X
Y2 = X**2
Y_log = np.log(X)
Y_cos = np.cos(np.pi*X)
```

## Intro Figures

```python
matplotlib.rcParams['font.size'] = 16
```

```python
plt.plot(X, Y2, X, Y, X, Y_log)
plt.xlabel('$x \in [1, \infty)$', fontsize=14)
plt.ylabel('$y$',fontsize=14)
plt.legend(['$y=x^2$', '$y=x$', '$y=\ln\;x$'], fontsize=12)
plt.title('Comparison of $\mathcal{O}(x^2)$, $\mathcal{O}(x)$, and $\mathcal{O}(\ln\;x)$', fontsize=16)
plt.text(3.4, 8, 'Rapid growth\nwhen $y \sim \mathcal{O}(x^2)$', fontsize=14, color='gray')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
```

## Symbols and Sub/Superscripts

Write symbols and sub/superscripts in matplotlib with $\TeX$.

**!!!** Be sure to use the 'r' prefix so that Python leaves the text as a raw string and does not convert escape sequences.

**Superscripts**

```python
plt.plot(X, Y2)
plt.title(r'$y=x^2$');
```

**Subscripts**

```python
plt.plot(X, Y)
plt.title(r'$y=x_i$');
```

```python
plt.plot(X, Y)
plt.title(r'$y=x_{base}$');   
```

**Symbols: Greek letters, Math functions**

```python
plt.plot(X, Y_cos)
plt.text(2.3, 0.75, r'$y=\cos(\pi x)$');  
```

```python
plt.plot(X, Y_cos)
plt.title(r'$y=\cos(\pi x)$')
plt.text(2.3, 0.75, r'$y=\cos(\pi x)$')
plt.xlabel(r'$y=\cos(\pi x)$')
plt.ylabel(r'$y=\cos(\pi x)$')
plt.legend([r'$y=\cos(\pi x)$'], loc=4); 
```

## Fractions and Formulas

$\TeX$ also allows you to create fractions and full equations.

**!!!** Be sure to use the 'r' prefix so that Python leaves the text as a raw string and does not convert escape sequences.

```python
x = np.linspace(0.5, 100, 1000)
y_frac = 2 + 1/(x**2);
```

**Fractions**

```python
plt.plot(x, y_frac)
plt.ylim(0, None)
plt.title(r'$y = 2 + \frac{1}{x^2}$', fontsize=20);  
```

```python
plt.plot(x, y_frac)
plt.ylim(0, None)
plt.text(40, 4, r'$\lim_{x\rightarrow\infty}\;2 + \frac{1}{x^2} = 2$', fontsize=20);
```
