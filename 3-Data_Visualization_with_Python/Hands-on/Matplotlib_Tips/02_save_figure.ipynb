# Matplotlib: Saving a Figure

```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
np.random.seed(99)

X = np.random.rand(100)
Y = 2*X + 0.5*np.random.rand(100)
```

```python
plt.scatter(X, Y);
```

## savefig

```python
plt.scatter(X, Y)
plt.savefig('random_scatter.png');
```

Matplotlib saves to **current directory** by default

Specify **file path** for other directory

```python
plt.scatter(X, Y)
plt.savefig('/Users/kimberlyfessel/Desktop/random_scatter.png');
```

## File Options - png, pdf, svg, ps

```python
plt.scatter(X, Y)
plt.savefig('random_scatter.svg');
```

```python
!ls
```

### Increase Resolution 

For higher resolution graphics, increase the **dots per inch (dpi)**.  Default dpi is 100.

```python
plt.scatter(X, Y)
plt.savefig('random_scatter_highres.png', dpi=200);
```

```python
!du -h random_scatter*.*
```

## Figure Cutting Off

```python
plt.scatter(X, Y)
plt.xlabel('Random Data', fontsize=24)
plt.savefig('random_scatter_xlabel.png');
```

```python
!open random_scatter_xlabel.png
```

```python
plt.scatter(X, Y)
plt.xlabel('Random Data', fontsize=24)
plt.savefig('random_scatter_xlabel_fixed.png', bbox_inches='tight');
```
