# Matplotlib: Changing Axes Limits

```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
X = np.linspace(1, 4, 1000)
Y = 2 + X**2
```

## pyplot xlim and ylim

Pass values to `plt.xlim()` or `plt.ylim()` to adjust the x- or y-axis limits.

```python
plt.plot(X, Y)
plt.xlim(0, 5);
```

```python
plt.plot(X, Y)
plt.ylim(0, 20);
```

```python
plt.plot(X, Y)
plt.xlim(0, 5)
plt.ylim(0, 20);
```

### Adjusting only one limit

```python
plt.plot(X, Y)
plt.ylim(0, None);
```

```python
plt.plot(X, Y)
plt.ylim(bottom=0);  #top, left, right also work
```

## axis set_xlim and set_ylim

Alternatively: pass values to `set_xlim()` or `set_ylim()` to specific axes.

```python
plt.plot(X, Y)

ax=plt.gca()
ax.set_xlim(0, 5)
ax.set_ylim(0, None);
```

```python
type(ax)
```

### Subplots

```python
plt.subplots(1, 3)

plt.subplot(1, 3, 1)
plt.plot(X, Y)

ax2 = plt.subplot(1, 3, 2)
plt.scatter(X[::100], Y[::100])

#or you could put the xlim, ylim functions 
#   here to change this subplot!

plt.subplot(1, 3, 3)
plt.plot(X, -Y)

plt.tight_layout()

plt.xlim(0, 5)
plt.ylim(-20, 0)

ax2.set_xlim(-2, 6)
ax2.set_ylim(-10, 25);
```

## BONUS: Reversing the axis

```python
plt.plot(X, Y)
plt.xlim(4, 1);
```

```python
plt.plot(X, Y)
left, right = plt.xlim()
plt.xlim(right, left);
```
