```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("planets")
```

```python
df
```

```python
df.groupby("method")[["distance"]].sum()
```

```python
df.groupby("method")[["distance"]].count()
```

```python
df.groupby("method")["distance"].value_counts()
```

```python
df.groupby("method")[["distance"]].mean()
```

```python
(df.groupby('method')['distance'].sum())/(df.groupby('method')['distance'].count())
```

```python
df.groupby("method")[["distance"]].max()
```

```python
df.groupby("method").distance.aggregate([min,np.median,max,np.sum,np.std])
```

```python
df.groupby("method").transform(lambda x: x.mean())
```

```python
df.groupby("method").distance.transform(lambda x: x*2)
```

```python
df.groupby("method").transform(lambda x: x*2)
```

```python
df.groupby("method").transform(lambda x: x)
```

```python
df.groupby("method").filter(lambda x: x['distance'].mean() < 100)

```
